```python
from ultralytics import YOLO
import shutil, os
```

```python
model=YOLO("yolov8n.pt")

results=model.train(
    data=dataset.location+"/data.yaml",
    epochs=15,
    imgsz=640,
    batch=16,
    workers=2,
    device=0
)

os.makedirs("saved_models",exist_ok=True)
shutil.copy("runs/detect/train/weights/best.pt","saved_models/baggage_best.pt")
shutil.copy("runs/detect/train/weights/last.pt","saved_models/baggage_plate_last.pt")

print("saved weights")
```
